```python
bottom = (0, 1, ("L", "J", "|"))
top = (0, -1, ("7", "F", "|"))
left = (-1, 0, ("L", "F", "-"))
right = (1, 0, ("7", "J", "-"))

symbols = {
    "|": (top, bottom),
    "-": (left, right),
    "L": (top, right),
    "J": (top, left),
    "7": (bottom, left),
    "F": (bottom, right),
}
```

```python
with open("input.txt") as f:
    data = f.read().splitlines()
lines = [list(line) for line in data]

x_size = len(lines[0])
y_size = len(lines)
```

```python
for y, line in enumerate(lines):
    if "S" in line:
        print(line.index("S"), y)
```

```python
visited = [
    (10, 61), (10, 60),
]

queue = set(
    [visited[-1]]
)

def visit_neighbours(x, y):
    symbol = lines[y][x]
    directions = symbols.get(symbol)
    for direction in directions:
        x_change, y_change, next_symbol = direction
        new_x = x + x_change
        new_y = y + y_change
        print(f"New pos : {new_x}/{new_y}", end=" ")
        
        if not(0 < new_x < x_size) and not(0 < new_y < y_size):
            print("Out of bounderies")
            continue
        
        if lines[new_y][new_x] in next_symbol:
            if (new_x, new_y) not in visited:
                visited.append((new_x, new_y))
                queue.add((new_x, new_y))
                print(f"Can continue to {new_x} : {new_y}")
                continue
            else:
                print("Already visited")
                continue
        
        print("Not a good symbol")
            
        
                
```

```python
visit_neighbours(*visited[-1])

while queue:
    visit_neighbours(*queue.pop())
```
