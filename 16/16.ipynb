```python
import copy

with open("input.txt") as f:
    data = [list(line) for line in f.read().splitlines()]
```

```python
def print_map(data):
    for line in data:
        line_content = "".join(line)
        print(f"{line_content}")
```

```python
print_map(data)
```

```python
from enum import Enum

class Directions(Enum):
    UP = 0
    DOWN = 1
    LEFT = 2
    RIGHT = 3

class Position:
    def __init__(self, x, y, direction):
        self.x = x
        self.y = y
        self.direction = direction

    def __repr__(self):
        return f"{self.x} / {self.y} / {self.direction}"

    def move(self):
        match self.direction:
            case Directions.UP:
                self.y -= 1
            case Directions.DOWN:
                self.y += 1
            case Directions.RIGHT:
                self.x += 1
            case Directions.LEFT:
                self.x -= 1
```

```python
p = Position(0, 0, Directions.RIGHT)
```

```python
output = [
    [" " for _ in range(len(data[0]))] for _ in range(len(data))
]
beam_counter = {}

moves = [Position(-1, 0, Directions.RIGHT)]

x_len = len(data[0])
y_len = len(data)

c = 0

while moves:
    p = moves.pop(0)

    p.move()

    if not( (0 <= p.x < x_len) & (0 <= p.y < y_len)):
        # print(f"{p.x}|{x_len} {p.y}|{y_len} NEW : Out of range")
        continue

    output[p.y][p.x] = "#"
    counter_key = f"{p.x} {p.y}"
    
    beam_counter.setdefault(counter_key, 0)
    beam_counter[counter_key] += 1

    # You cannot have more than 4 times the beam on a cell
    if beam_counter[counter_key] > 4:
        continue
    
    match data[p.y][p.x]:
        case ".":
            moves.append(Position(p.x, p.y, p.direction))
        case "|":
            if p.direction in [Directions.LEFT, Directions.RIGHT]:
                moves += [
                    Position(p.x, p.y, Directions.UP),
                    Position(p.x, p.y, Directions.DOWN),
                ]
            else:
                moves.append(p)
                    
        case "-":
            if p.direction in [Directions.UP, Directions.DOWN]:
                moves += [
                    Position(p.x, p.y, Directions.LEFT),
                    Position(p.x, p.y, Directions.RIGHT),
                ]
            else:
                moves.append(p)
        case "/":
            match p.direction:
                case Directions.UP:
                    moves.append(Position(p.x, p.y, Directions.RIGHT))
                case Directions.DOWN:
                    moves.append(Position(p.x, p.y, Directions.LEFT))
                case Directions.LEFT:
                   moves.append(Position(p.x, p.y, Directions.DOWN))
                case Directions.RIGHT:
                    moves.append(Position(p.x, p.y, Directions.UP))   
        case "\\":
            match p.direction:
                case Directions.UP:
                    moves.append(Position(p.x, p.y, Directions.LEFT))
                case Directions.DOWN:
                    moves.append(Position(p.x, p.y, Directions.RIGHT))
                case Directions.LEFT:
                    moves.append(Position(p.x, p.y, Directions.UP))
                case Directions.RIGHT:
                    moves += [
                        Position(p.x, p.y, Directions.DOWN),
                    ]   
print(c)
```

```python
print_map(output)
```
