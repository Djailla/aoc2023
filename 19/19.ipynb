```python
from itertools import groupby

with open("input.txt") as f:
    data = f.read().splitlines()

rules, positions = [list(group) for key, group in groupby(data, lambda x: x == "") if not key]

rule_dict = {}

for r in rules:
    r_name, sub_rules = r[:-1].split("{")
    rule_dict[r_name] = []

    for sub_rule in sub_rules.split(","):
        if ":" in sub_rule:
            rule_test, rule_res = sub_rule.split(":")
            rule_key = rule_test[0]
            rule_op = rule_test[1]
            rule_val = int(rule_test[2:])
            
            rule_dict[r_name].append((rule_key, rule_op, rule_val, rule_res))
        else:
            rule_dict[r_name].append((sub_rule, ))

class Position:
    def __init__(self, x, m, a, s):
        self.x = x
        self.m = m
        self.a = a
        self.s = s

        self.accepted = False

    def __repr__(self):
        return f"{self.x=} {self.m=} {self.a=} {self.s=} => {self.accepted}"

    @property
    def score(self):
        return self.x + self.m + self.a + self.s
    
    def test_rules(self, rules):
        loop = 0
        print(f"Start rules = {rules}")

        while rules:
            loop += 1
            if loop > 100:
                break

            print(f"Loop rules = {rules}")
            
            rule = rules.pop(0)
            
            if len(rule) == 1:
                rule = rule[0]
                match rule:
                    case "A":
                        print("ACCEPTED\n\n\n")
                        self.accepted = True
                        break
                    case "R":
                        print("REJECTED\n\n\n")
                        break
                    case _:
                        print(rule_dict[rule])
                        rules = rule_dict[rule].copy() + rules
    
            if len(rule) == 4:
                rule_key, rule_op, rule_val, rule_next = rule
                pos_val = getattr(self, rule_key)
                print(pos_val)
                if rule_op == ">":
                    res = pos_val > rule_val
                elif rule_op == "<":
                    res = pos_val < rule_val
    
                print(f"{pos_val} {rule_op} {rule_val} = {res}")
                if res:
                    match rule_next:
                        case "A":
                            print("ACCEPTED\n\n\n")
                            self.accepted = True
                            break
                        case "R":
                            print("REJECTED\n\n\n")
                            break
                        case _:
                            print(rule_dict[rule_next])
                            rules = rule_dict[rule_next].copy() + rules
                else:
                    print("FAILED TEST GO TO NEXT RULE")
                    continue

        print("LOOP DONE\n\n\n")
```

```python
test_pos = [
    Position(*[
        int(val[2:]) for val in pos[1:-1].split(",")
    ]) for pos in positions
]
```

```python
for pos in test_pos:
    pos.test_rules(rule_dict["in"].copy())
```

```python
sum(pos.score for pos in test_pos if pos.accepted)
```
